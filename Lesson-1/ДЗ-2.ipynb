1.​ ​Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)

1) Найдите их сумму. (на листочке)

2) Напишите код на Python, реализующий расчет длины вектора, заданного его
координатами. (в программе)

```python
import numpy as np
import math
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

a = np.array([20, 20, 20])
b = np.array([0, 0, -20])
c = a + b
print(f'Сумма векторов а и b = {c}')
def vector_length(vect):
    l = math.sqrt(vect[0]**2 + vect[1]**2 + vect[2]**2)
    print(f'Длинна вектора , заданного его координатами {vect}: {l}')
vector_length(a)
vector_length(b)
vector_length(c)
```

3.​ ​Задание (в программе)
Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

```python
#график окружности
plt.figure(figsize=(5, 5))
x = []
y1 = []
y2 = []
r = 10
for i in range(-1000, 1001):
    x_ = i/100
    x.append(x_)
    y1.append(math.sqrt(r**2 - x_**2))
    y2.append(-math.sqrt(r**2 - x_**2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#график эллипса
plt.figure(figsize=(5, 5))
x = []
y1 = []
y2 = []
a = 3
b = 2
for i in range(-300, 301):
    x_ = i/100
    x.append(x_)
    y1.append(math.sqrt((a**2*b**2-x_**2*b**2)/a**2))
    y2.append(-math.sqrt((a**2*b**2-x_**2*b**2)/a**2))
plt.ylim(-3, 3)
plt.plot(x,y1);
plt.plot(x,y2);
plt.xlabel('x')
plt.ylabel('y')
```

```python
#график гиперболы
plt.figure(figsize=(5, 5))
x = []
y1 = []
y2 = []
a = 2
b = 4
for i in range(1000):
    x_ = a + i/100
    x.append(x_)
    y1.append(-math.sqrt(x_**2/a**2 - 1) * b)
    y2.append(math.sqrt(x_**2/a**2 - 1) * b)
    
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.plot(x,y1);
plt.plot(x,y2);
plt.xlabel('x')
plt.ylabel('y')
```

5. Задание (в программе)
1) Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)

X, Y = np.meshgrid(X, Y)
Z1 = 3*Y - 2*X - 5
Z2 = 3*Y - 2*X + 3
ax.plot_surface(X, Y, Z1) 
ax.plot_surface(X, Y, Z2) 
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```

2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

Гипералический параболоид

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
a = 5
b = 8
X, Y = np.meshgrid(X, Y)
Z = X**2/a**2 - Y**2/b**2
ax.plot_surface(X, Y, Z)
show()
```

Однополюсный гиперболоид

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
a = 5
b = 8
с = 20
X, Y = np.meshgrid(X, Y)
Z = (1 - (X**2/a**2 + Y**2/b**2))/с**2
ax.plot_surface(X, Y, Z)
show()
```
