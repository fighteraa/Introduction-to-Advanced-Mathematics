1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-7, 7, 210)
k1 = 1
k2 = 3.5
k3 = -3.5
a1 = 2
a2 = -2
a3 = -1
b1 = -0.5
b2 = -0.5
b3 = 1.5
y1 = k1*np.cos(x-a1) + b1
y2 = k2*np.cos(x-a2) + b2
y3 = k3*np.cos(x-a3) + b3
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.
Напишите код, который будет рисовать график окружности в полярных координатах.

```python
R = 10
angle = np.linspace(0, 2*np.pi, 1000)

x = R * np.cos(angle)
y = R * np.sin(angle)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# График окружности с единичным радиусом
alpha = np.linspace(0, np.pi, 1000)
x = np.cos(alpha)
y = np.sin(alpha)
R = np.sqrt(x**2 + y**2)
angle = np.arctan2(y, x)*180/np.pi
plt.polar(angle, R);
```

4. Задание (в программе)
Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-2, 7, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 30)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y -1)
    
x1, y1 = fsolve(equations, (-2, 1))
print(x1, y1)

x2, y2 = fsolve(equations, (2, 1))
print(x2, y2)

x3, y3 = fsolve(equations, (5, 25))
print(x3, y3)
```
